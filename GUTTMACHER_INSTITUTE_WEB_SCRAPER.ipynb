```python
import pandas as pd
import numpy as np
import os

!mkdir GUTTMACHER

name_list = os.listdir("GUTTMACHER\\")

name_list

if ".ipynb_checkpoints" in name_list:
    name_list.remove(".ipynb_checkpoints")
if "desktop.ini" in name_list:
    name_list.remove("desktop.ini")

data = []
for k in range(len(name_list)):
    x = "GUTTMACHER\\" + name_list[k]
    df = pd.read_csv(x)
    data.append(df)
data=pd.concat(data)

table = pd.pivot_table(data, values='datum', index=['county_id'],
                    columns=['measure_name'], aggfunc=np.sum)

table.to_csv("abortion_clinics.csv")
```
